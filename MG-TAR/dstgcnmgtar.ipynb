```python
import os
import json
import torch
import dgl
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
from dgl.nn.pytorch import GraphConv
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr
from tqdm.auto import tqdm
import torch.nn.functional as F

# Setup Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
warnings.filterwarnings('ignore')

# ==============================================================================
# 1. DATA LOADER (ĐÃ SỬA LỖI DATA SPLIT)
# ==============================================================================
def data_loader(data_path, city, year, level='district', length=12, n_steps=12, is_scale=True, train_ratio=0.8, temporal_copy=True):
    
    def normalize(train, test):
        if is_scale:
            scaler = MinMaxScaler(feature_range=(0, 1))
            # Flatten [Samples, Nodes, Feats] -> [Samples*Nodes, Feats] để fit
            train_shape = train.shape
            test_shape = test.shape
            train_flat = train.reshape(-1, train_shape[-1])
            test_flat = test.reshape(-1, test_shape[-1])
            
            train_norm = scaler.fit_transform(train_flat)
            test_norm = scaler.transform(test_flat)
            
            return train_norm.reshape(train_shape), test_norm.reshape(test_shape), scaler
        else:
            return train, test, None

    # --- 1. Load Risk Data (Target) ---
    print(f"Loading Risk Data for {city}-{year}...")
    risk_df = pd.read_csv(f'{data_path}/risk_scores/{city}-{year}-{level}-hour-risk.csv')
    selected_areas = risk_df.drop(columns=['date', 'time']).columns
    n_districts = len(selected_areas)
    
    # FIX: Tính điểm cắt dựa trên tổng dữ liệu thực tế
    total_samples = risk_df.shape[0]
    split_idx = int(total_samples * train_ratio)
    print(f"-> Total samples: {total_samples}. Splitting at index: {split_idx}")

    # Prepare Risk Arrays
    risk_vals = risk_df[selected_areas].values
    
    # Tạo Sliding Window
    X_risk, Y_risk = [], []
    # Loop từ length đến total - n_steps
    # Input: [i-length : i], Output: [i : i+n_steps] (Lấy step đầu làm target chính)
    
    limit = total_samples - n_steps
    for i in range(length, limit):
        X_risk.append(risk_vals[i-length : i])      # [Length, Nodes]
        Y_risk.append(risk_vals[i : i+n_steps])     # [n_steps, Nodes]
        
    X_risk = np.array(X_risk) # [Samples, Length, Nodes]
    Y_risk = np.array(Y_risk) # [Samples, n_steps, Nodes]
    
    # Split Train/Test
    # Cần tính lại index cắt cho mảng đã slide (trừ đi length offset)
    real_split = split_idx - length
    
    risk_train = X_risk[:real_split]
    risk_test = X_risk[real_split:]
    y_train = Y_risk[:real_split]
    y_test = Y_risk[real_split:]
    
    # Normalize Risk
    risk_train, risk_test, risk_scaler = normalize(risk_train, risk_test)
    y_train_scaled, y_test_scaled, _ = normalize(y_train, y_test)
    
    # Reshape cho đúng format model [Samples, Length, Nodes, 1]
    risk_train = risk_train[..., None]
    risk_test = risk_test[..., None]

    # --- 2. Helper Load Features ---
    def load_feature(folder_name, file_pattern, is_static=False):
        # Hàm load tổng quát để code gọn hơn
        # Logic: Load CSV -> Filter Columns -> Stack -> Slide -> Split -> Norm
        if folder_name == 'weather':
            df = pd.read_csv(f'{data_path}/weather/{city}-{year}-count.csv').fillna(0)
            if level == 'district':
                df['location'] = df['location'].apply(lambda x: x.split('|')[0])
                df = df.groupby(['date','time','location'], as_index=False).mean()
            # Extract per location
            feats = []
            for loc in selected_areas:
                feats.append(df[df['location'] == loc].iloc[:, 3:].values)
            full_data = np.stack(feats, axis=1) # [Time, Nodes, Feats]
            
        elif folder_name == 'dangerous_cases': # DTG
            df = pd.read_csv(f'{data_path}/dangerous_cases/{city}-{year}-date-hour-{level}-new.csv')
            feats = []
            for loc in selected_areas:
                dist = loc.split('|')[0]
                feats.append(df[df['district'] == dist].iloc[:, 3:].values)
            full_data = np.stack(feats, axis=1)

        elif folder_name in ['traffic_volume', 'traffic_speed']:
            fname = 'traffic_volume' if folder_name == 'traffic_volume' else 'traffic_speed'
            df = pd.read_csv(f'{data_path}/{folder_name}/{city}-{year}.csv').drop(columns=['date','hour'])
            full_data = df[selected_areas].values[..., None] # [Time, Nodes, 1]
            
        elif folder_name == 'calendar':
            df = pd.read_csv(f'{data_path}/calendar/calendar-{city}-{year}-{level}.csv')
            feats = [df[df['location'] == loc].iloc[:, 1:].values for loc in selected_areas]
            full_data = np.stack(feats, axis=1)
            
        # Sliding Window
        X_feat = []
        # Cắt cho khớp với Risk (Risk lấy từ length -> limit)
        # Feature input cũng lấy [i-length : i]
        # Lưu ý: full_data có thể dài hơn limit, chỉ lấy đủ dùng
        for i in range(length, limit):
            if i > len(full_data): break
            X_feat.append(full_data[i-length : i])
            
        X_feat = np.array(X_feat)
        
        # Split
        train_f = X_feat[:real_split]
        test_f = X_feat[real_split:]
        
        # Norm
        train_f, test_f, _ = normalize(train_f, test_f)
        return train_f, test_f

    # Load Dynamic Features
    print("Loading Dynamic Features...")
    weather_train, weather_test = load_feature('weather', '')
    dtg_train, dtg_test = load_feature('dangerous_cases', '')
    vol_train, vol_test = load_feature('traffic_volume', '')
    spd_train, spd_test = load_feature('traffic_speed', '')
    cal_train, cal_test = load_feature('calendar', '')
    
    # Load Static Features
    print("Loading Static Features...")
    road = pd.read_csv(f'{data_path}/roads/{city}-{year}-{level}-road-count.csv').drop(columns=['attribute'])[selected_areas].T.values
    demo = pd.read_csv(f'{data_path}/demographic/{city}-{year}-{level}.csv').drop(columns=['index'])[selected_areas].T.values
    poi = pd.read_csv(f'{data_path}/poi/{city}-{year}-{level}.csv').drop(columns=['location'])[selected_areas].T.values
    
    scaler_s = MinMaxScaler()
    spatial_static = np.concatenate([
        scaler_s.fit_transform(road),
        scaler_s.fit_transform(demo),
        scaler_s.fit_transform(poi)
    ], axis=-1)

    print(f"-> Train samples: {len(risk_train)} | Test samples: {len(risk_test)}")
    
    return {
        'risk': [risk_train, risk_test],
        'volume': [vol_train, vol_test],
        'speed': [spd_train, spd_test],
        'dtg': [dtg_train, dtg_test],
        'weather': [weather_train, weather_test],
        'calendar': [cal_train, cal_test],
        'spatial_static': spatial_static,
        'y': [y_train, y_test], # Raw Target
        'y_scaled': [y_train_scaled, y_test_scaled], # Scaled Target
        'selected_areas': selected_areas,
        'scaler': risk_scaler
    }

# ==============================================================================
# 2. DATASET WRAPPER
# ==============================================================================
class SeoulRiskDataset(Dataset):
    def __init__(self, data_dict, mode='train', graph=None):
        self.idx = 0 if mode == 'train' else 1
        self.graph = graph
        self.d = data_dict
        self.len = self.d['risk'][self.idx].shape[0]
        
    def __len__(self): return self.len
    
    def __getitem__(self, i):
        # 1. Temporal: [T, N, F] -> Transpose to [N, T, F]
        # Kết hợp Risk, Vol, Speed, DTG
        dyn = []
        for key in ['risk', 'volume', 'speed', 'dtg']:
            # Data gốc: [Samples, T, N, F] -> Lấy sample i: [T, N, F]
            # Transpose -> [N, T, F]
            dyn.append(self.d[key][self.idx][i].transpose(1,0,2))
        temporal = np.concatenate(dyn, axis=-1)
        
        # 2. External: Weather, Calendar -> Mean -> [N, F]
        ext = []
        for key in ['weather', 'calendar']:
            ext.append(self.d[key][self.idx][i].transpose(1,0,2))
        external = np.mean(np.concatenate(ext, axis=-1), axis=1)
        
        # 3. Static
        spatial = self.d['spatial_static']
        
        # 4. Target (Lấy step đầu tiên để dự đoán)
        # y_scaled: [Samples, n_steps, Nodes] -> [Nodes, 1]
        y_s = self.d['y_scaled'][self.idx][i][0][:, None]
        y_r = self.d['y'][self.idx][i][0][:, None]
        
        return (self.graph, 
                torch.FloatTensor(spatial), 
                torch.FloatTensor(temporal), 
                torch.FloatTensor(external),
                torch.FloatTensor(y_s),
                torch.FloatTensor(y_r))

def collate_fn(batch):
    g, s, t, e, ys, yr = zip(*batch)
    return dgl.batch(g), torch.cat(s), torch.cat(t), torch.cat(e), torch.cat(ys), torch.cat(yr)

# ==============================================================================
# 3. MODEL ARCHITECTURE (FIXED)
# ==============================================================================
class GCN(nn.Module):
    def __init__(self, in_feats, out_feats):
        super(GCN, self).__init__()
        self.conv = GraphConv(in_feats, out_feats, allow_zero_in_degree=True)
        self.bn = nn.BatchNorm1d(out_feats)
        
    def forward(self, g, h):
        h = self.conv(g, h)
        # Handle Batch Norm for Graph (flattened nodes)
        h = self.bn(h)
        return F.relu(h)

class DSTGCN_V2(nn.Module):
    def __init__(self, f_spatial, f_temporal, f_external):
        super(DSTGCN_V2, self).__init__()
        
        HIDDEN = 64
        
        # Spatial Branch
        self.s_enc = nn.Linear(f_spatial, HIDDEN)
        self.s_gcn = GCN(HIDDEN, 32)
        
        # Temporal Branch
        # Input: [N, T, F_t] -> Transpose [N, F_t, T] for Conv1d
        self.t_conv1 = nn.Conv1d(f_temporal, 32, kernel_size=3, padding=1)
        self.t_conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        
        self.t_agg = nn.Sequential(
            nn.Linear(64, 32),
            nn.ReLU()
        )
        
        # External Branch
        self.e_fc = nn.Linear(f_external, 16)
        
        # Fusion & Output (Regression)
        # 32 (Spatial) + 32 (Temporal) + 16 (External) = 80
        self.head = nn.Sequential(
            nn.Linear(80, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 1) # No Sigmoid -> Linear Output
        )
        
    def forward(self, g, x_s, x_t, x_e):
        # Spatial
        s = F.relu(self.s_enc(x_s))
        s_out = self.s_gcn(g, s)
        
        # Temporal
        # x_t: [N, T, F] -> [N, F, T]
        t = x_t.transpose(1, 2)
        t = F.relu(self.t_conv1(t))
        t = F.relu(self.t_conv2(t))
        t_mean = torch.mean(t, dim=2) # Pooling over time -> [N, 64]
        t_out = self.t_agg(t_mean)
        
        # External
        e_out = F.relu(self.e_fc(x_e))
        
        # Fusion
        fused = torch.cat([s_out, t_out, e_out], dim=1)
        return self.head(fused)

# ==============================================================================
# 4. WEIGHTED LOSS (Để xử lý dữ liệu nhiều số 0)
# ==============================================================================
class WeightedMSELoss(nn.Module):
    def __init__(self, weight=10.0):
        super().__init__()
        self.weight = weight
        
    def forward(self, pred, target):
        loss = (pred - target) ** 2
        # Nếu target > 0 (có tai nạn), nhân loss lên weight lần
        mask = target > 0.01 
        w_tensor = torch.ones_like(loss)
        w_tensor[mask] = self.weight
        return torch.mean(loss * w_tensor)

# ==============================================================================
# 5. MAIN EXECUTION
# ==============================================================================
def main():
    # --- CONFIG ---
    CITY, YEAR = 'Seoul', '2016' # Hoặc 2018
    DATA_PATH = '/kaggle/input/mg-tar'
    BATCH_SIZE = 64
    EPOCHS = 50
    LR = 0.003
    
    # 1. Load Data
    data = data_loader(DATA_PATH, CITY, YEAR, is_scale=True, train_ratio=0.8)
    N_NODES = len(data['selected_areas'])
    
    # 2. Graph
    adj_path = f'{DATA_PATH}/graph_data/{CITY}-normalized-district.csv'
    adj = pd.read_csv(adj_path, index_col=0).values
    src, dst = np.nonzero(adj)
    g = dgl.graph((src, dst)).to(device)
    g = dgl.add_self_loop(g)
    
    # 3. Loader
    train_ds = SeoulRiskDataset(data, 'train', g)
    test_ds = SeoulRiskDataset(data, 'test', g)
    train_loader = DataLoader(train_ds, BATCH_SIZE, shuffle=True, collate_fn=collate_fn)
    test_loader = DataLoader(test_ds, BATCH_SIZE, shuffle=False, collate_fn=collate_fn)
    
    # 4. Init Model
    s0 = train_ds[0]
    model = DSTGCN_V2(
        f_spatial=s0[1].shape[1],
        f_temporal=s0[2].shape[2],
        f_external=s0[3].shape[1]
    ).to(device)
    
    optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=1e-4)
    # Weighted Loss: Phạt gấp 20 lần nếu sai ở chỗ có tai nạn
    criterion = WeightedMSELoss(weight=20.0) 
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=5, verbose=True)
    
    # 5. Training
    print("\nStarting Training...")
    best_loss = float('inf')
    
    for epoch in range(EPOCHS):
        model.train()
        train_loss = 0
        for bg, s, t, e, ys, _ in tqdm(train_loader, desc=f"Ep {epoch+1}", leave=False):
            bg, s, t, e, ys = bg.to(device), s.to(device), t.to(device), e.to(device), ys.to(device)
            
            optimizer.zero_grad()
            pred = model(bg, s, t, e)
            loss = criterion(pred, ys)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            train_loss += loss.item()
            
        avg_train = train_loss / len(train_loader)
        
        # Validation
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for bg, s, t, e, ys, _ in test_loader:
                bg, s, t, e, ys = bg.to(device), s.to(device), t.to(device), e.to(device), ys.to(device)
                pred = model(bg, s, t, e)
                val_loss += criterion(pred, ys).item()
        
        avg_val = val_loss / len(test_loader)
        scheduler.step(avg_val)
        
        print(f"Epoch {epoch+1}: Train Loss: {avg_train:.5f} | Val Loss: {avg_val:.5f}")
        
        if avg_val < best_loss:
            best_loss = avg_val
            torch.save(model.state_dict(), 'dstgcn_final.pth')
            
    # 6. Final Evaluation
    # 6. Final Evaluation
    print("\n--- Final Evaluation ---")
    model.load_state_dict(torch.load('dstgcn_final.pth'))
    model.eval()
    
    all_preds, all_trues = [], []
    scaler = data['scaler']
    
    with torch.no_grad():
        for bg, s, t, e, _, yr in tqdm(test_loader, desc="Testing"):
            bg, s, t, e = bg.to(device), s.to(device), t.to(device), e.to(device)
            
            # Predict
            pred_s = model(bg, s, t, e).cpu().numpy()
            
            # Inverse Scale
            # Reshape [Batch*Nodes, 1] -> [Batch, Nodes] -> Inverse -> Flatten
            try:
                pred_reshaped = pred_s.reshape(-1, N_NODES)
                pred_real = scaler.inverse_transform(pred_reshaped).flatten()
            except ValueError:
                continue # Skip last batch if mismatch
                
            all_preds.append(pred_real)
            all_trues.append(yr.numpy().flatten())
            
    y_pred = np.concatenate(all_preds)
    y_true = np.concatenate(all_trues)
    
    # Clip prediction < 0 (Risk không thể âm)
    y_pred = np.maximum(y_pred, 0)
    
    # Metrics
    mse = mean_squared_error(y_true, y_pred)     # <--- Đã thêm MSE
    rmse = np.sqrt(mse)                          # Tính RMSE từ MSE
    mae = mean_absolute_error(y_true, y_pred)
    pcc = pearsonr(y_true, y_pred)[0]
    
    # Check PCC on High Risk Points
    mask = y_true > 0
    pcc_risk = pearsonr(y_true[mask], y_pred[mask])[0] if mask.sum() > 0 else 0
    
    print("-" * 30)
    print(f"MSE         : {mse:.4f}")            # <--- In ra MSE
    print(f"RMSE        : {rmse:.4f}")
    print(f"MAE         : {mae:.4f}")
    print(f"PCC (All)   : {pcc:.4f}")
    print(f"PCC (Risk>0): {pcc_risk:.4f}")
    print("-" * 30)

    # Visualization
    plt.figure(figsize=(15, 6))
    plt.plot(y_true[:200], label='Ground Truth', color='black', alpha=0.6)
    plt.plot(y_pred[:200], label='Prediction', color='red')
    plt.title(f"Prediction Sample - PCC: {pcc:.4f}")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
